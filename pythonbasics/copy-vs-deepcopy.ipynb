```python
import copy
```

```python
# initialize list one
l1 = [1,2, [3,5], 4]
```

```python
# deep copy
l2 = copy.deepcopy(l1)
```

```python
# deep copy
l2
```

```python
l2[2]
```

```python
# Modifications in list 2 will not reflect in list as it is a deep copy 
l2[2][0] = 100
```

```python

l2
```

```python
l1
```

```python
# where as in shallow copy a reference of object is copied in other object.
# It means that any changes made to a copy of object do reflect in the original object.
s1 = [1,2,[3,4],5]
```

```python
s2 = copy.copy(s1)
```

```python
s2[2][0] = 500
```

```python
s1
```

```python
s2
```
